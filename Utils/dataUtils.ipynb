```python
#æœˆæ¶ˆè´¹
def getMonthPurchase(df_feature,group='month_lag',fea='purchase_amount',name='before'):
    df_purchase = df_feature.groupby(['card_id','month_lag'])['purchase_amount'].sum().reset_index()
    df_purchase.rename(columns={'purchase_amount':'purchase_amount_sum_month'},inplace=True)
    df_purchase.sort_values(by=['month_lag'],inplace=True,ascending=True)
    df_count = df_purchase.groupby(['card_id'])['purchase_amount_sum_month'].count().reset_index()
    df_count.rename(columns={'purchase_amount_sum_month':'%s_purchase_counts'%name},inplace=True)

    df_temp = df_purchase[['card_id','month_lag','purchase_amount_sum_month']]
    df_temp.index = df_temp.card_id
    df_temp = df_temp.set_index(['month_lag'],append=True)
    df_temp = pd.Series(df_temp['purchase_amount_sum_month'].values.reshape(len(df_temp['purchase_amount_sum_month'])),index=df_temp.index)
    df_temp = df_temp.unstack()
    df_temp.reset_index(inplace=True)
    cols = ['card_id']
    for index in list(df_purchase['month_lag'].unique()):
        cols.append('%s_monthLag%s_purchase'%(name,index))
    df_temp.columns = cols
    df_temp = df_temp.merge(df_count,on='card_id',how='left')
    df_temp['%s_purchase_mean'%name] = 0
    df_temp.fillna(0,inplace=True)

    for index in list(df_purchase['month_lag'].unique()):
        df_temp['%s_purchase_mean'%name] +=df_temp['%s_monthLag%s_purchase'%(name,index)]
    df_temp['%s_purchase_mean'%name] = (df_temp['%s_purchase_mean'%name] - df_temp['%s_monthLag0_purchase'%name])/df_temp['%s_purchase_counts'%name]
    df_temp['base_diff_%s_purchase'%name] = df_temp['%s_monthLag0_purchase'%name] - df_temp['%s_purchase_mean'%name]
    df_temp['base_diff_%s_purchaseRatio'%name] = df_temp['base_diff_%s_purchase'%name]/df_temp['%s_purchase_mean'%name]
    return df_temp

df_temp = getMonthPurchase(df_historical)
```
